```python
from tensorflow.keras.layers import TextVectorization
```

```python
df.columns
```

```python
X = df['comment_text']
```

```python
X.head()
```

```python
Y = df.drop(columns=['comment_text', 'id'])
```

```python
Y.values
```

```python
max_words_in_vocabular = 200000
```

```python
vectorizer = TextVectorization(max_tokens=max_words_in_vocabular , output_sequence_length= 1800, output_mode='int')
```

```python
vectorizer.adapt(X.values)
```

```python
vectorizer.get_vocabulary()
```

```python
vectorizer('who is james bond?')
```

```python
vectorized_text = vectorizer(X.values)
```

```python
vectorized_text
```

```python
X.shape
```

```python
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text,Y))
dataset = dataset.cache()
dataset = dataset.shuffle(160000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8)
```

```python
batch1 , y_of_batch1 =dataset.as_numpy_iterator().next()
```

```python
batch1
```

```python
batch1.shape
```

```python
y_of_batch1
```

```python
train_set = dataset.take(int(len(dataset)*0.7))
val_set = dataset.skip(int(len(dataset)*0.7)).take(int(len(dataset)*0.2))
test_set = dataset.skip(int(len(dataset)*0.9)).take(int(len(dataset)*0.1))
```

```python
len(test_set)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM , Bidirectional , Dropout , Dense  ,Embedding
```

```python
model= Sequential()
```

```python
model.add(Embedding (max_words_in_vocabular +1, 32))
model.add(Bidirectional(LSTM(32 , activation = 'tanh')))
model.add(Dense(128 , activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(6 , activation='sigmoid'))
```

```python
model.compile(loss='BinaryCrossentropy' , optimizer = 'Adam')
```

```python
model.summary()
```

```python
history = model.fit(train_set, epochs = 5 ,validation_data=val_set)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
pd.DataFrame(history.history).plot()
plt.show()
```

```python
from tensorflow.keras.metrics import Precision , Recall , CategoricalAccuracy
```

```python
prec = Precision()
rec = Recall()
acc = CategoricalAccuracy()
```

```python
for batch in test_set.as_numpy_iterator():
    X , y = batch
    y_pred = model.predict(X)
    y = y.flatten()
    y_pred = y_pred.flatten()
    
    prec.update_state(y , y_pred)
    rec.update_state(y , y_pred)
    acc.update_state(y , y_pred)
```

```python
print(f'Precision: {prec.result().numpy()}, Recall:{rec.result().numpy()}, Accuracy:{acc.result().numpy()}')
```
